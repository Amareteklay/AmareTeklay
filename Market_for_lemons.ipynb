# Replicating Akerlof's paper on Market for lemons

```python
import matplotlib.pyplot as plt
```

```python
mu = 1.5
Y1 = 100
Y2 = 120
N = 10

def plot_demand(p):
    D1_values = []
    D2_values = []
    D_values = []
    for val in p:
        if mu / val > 1:
            D1 = Y1 / val  # Divide Y1 by each value in p
        else:
            D1 = 0
        D1_values.append(D1)
        if 3*mu / 2 > val:
            D2 = Y2 / val  # Divide Y1 by each value in p
        else:
            D2 = 0
        D2_values.append(D2)

        if mu > val:
          D = (Y2 + Y1) / val  # Divide Y1 by each value in p
        elif mu <= val < 3*mu/2:
          D = Y2/val
        else:
          D = 0
        D_values.append(D)
    return D1_values, D2_values, D_values

def plot_supply(p):
    S1_values = []
    S2_values = []
    for val in p:
      if val <= 2:
        S1 = val*N/2
      else:
        S1 = 0

      S1_values.append(S1)
      if val > 1:
        S2 = N
      else:
        S2 = 0
      S2_values.append(S2)
    return S1_values, S2_values

p_values = [i for i in range(1, 11)]  # Example input values from 1 to 10

demand_values = plot_demand(p_values)
supply_values = plot_supply(p_values)
plt.plot(p_values, demand_values[0], label=f'mu = {mu}')
plt.plot(p_values, demand_values[1])
plt.plot(p_values, demand_values[2])
plt.plot(p_values, supply_values[0], label = 'S1')
plt.plot(p_values, supply_values[1], label = 'S2')

plt.xlabel('Input (p)')
plt.ylabel('Demand (D1)')
plt.title('Demand as a function of Input for mu = 10')
plt.legend()
plt.grid(True)
plt.show()
```
